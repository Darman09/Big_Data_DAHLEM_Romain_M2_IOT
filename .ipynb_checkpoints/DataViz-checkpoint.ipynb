```python
# En crÃ©ant un nuage de points, regardez comment se comporte la colonne LotArea par rapport au SalesPrices 
g = sns.scatterplot(x="LotArea", y="SalePrice", data=dfExplo)
```

```python
# Affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotArea infÃ©rieur Ã  20 000 pieds carrÃ©s et un prix infÃ©rieur Ã  500 000$
dfMini = dfExplo[dfExplo['LotArea'] < 20000]
dfMini = dfMini[dfMini['SalePrice'] < 500000]
g = sns.scatterplot(x="LotArea", y="SalePrice", data=dfMini)
```
