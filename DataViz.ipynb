```python
'''
Visualisation d’un nuage de points et exploration des variables
'''

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ouverture du fichier 
dfExplo = pd.read_csv("data/house_pricing.csv",",",encoding='utf-8', low_memory=False)

dfExplo.info()
```

```python
# En créant un nuage de points, regardez comment se comporte la colonne LotArea par rapport au SalesPrices 
g = sns.scatterplot(x="LotArea", y="SalePrice", data=dfExplo)
```

```python
# Affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotArea inférieur à 20 000 pieds carrés et un prix inférieur à 500 000$
dfMini = dfExplo[dfExplo['LotArea'] < 20000]
dfMini = dfMini[dfMini['SalePrice'] < 500000]
g = sns.scatterplot(x="LotArea", y="SalePrice", data=dfMini)
```

```python
# En créant un nuage de points, regardez la relation entre le LotFrontage et le LotArea
g = sns.scatterplot(x="LotArea", y="LotFrontage", data=dfExplo)
```

```python
# De la même manière, affinez votre visualisation en ne gardant uniquement les maisons qui ont un LotFrontage inférieur à 200 pieds carrés et un LotArea inférieur à 100000 pieds carré
dfMini = dfExplo[dfExplo['LotArea'] < 100000]
dfMini = dfMini[dfMini['LotFrontage'] < 200]
g = sns.scatterplot(x="LotArea", y="LotFrontage", data=dfMini)



```
